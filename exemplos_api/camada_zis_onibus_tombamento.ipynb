```python
import os
import sys

#path para a biblioteca do apiModulo. Ajuste de acordo com a necessidade
sys.path.insert(0, os.path.abspath('../'))
from apiModulo.api import *

zis = lerArquivo('../../exemplos/zis/ZIS_1992_SIRGAS2000.shp')
```

```python
print(zis.crs)
zis = zis.set_crs(epsg=31983)
gdf = zis.to_crs(epsg=4326)
print(zis.crs)
print(gdf.crs)
```

```python
visMapaGJson(gdf, 
                    variavel='NOME', 
                    descricao='NOME',                     
                    height='800', 
                    width='90%')
```

```python
gdf.head()
```

```python
gdf.drop(['FID_','Entity', 'Layer',  'Color','Linetype', 'Elevation', 'LineWt', 'RefName' ],axis=1, inplace=True)
gdf.columns
```

```python
gdf = gdf.reset_index()
gdf['cod_camada'] = gdf.index
```

```python

inserirCamada(
    dado = gdf,
    tabela= 'zis',
    campo_chave='cod_camada',
    nome='zis',
    descricao='ZIS_1992_SIRGAS2000'
)
```
